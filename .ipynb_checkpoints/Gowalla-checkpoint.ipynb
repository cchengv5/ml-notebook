```python
import geohash
import pandas as pd
import numpy as np
import math


def merc(lat, lon):
    r_major = 6378137.00
    x = r_major * math.radians(lon)
    scale = x / lon
    y = 180.0 / math.pi * math.log(math.tan(math.pi / 4.0 + lat * (math.pi / 180.0) /2.0 )) * scale
    
    return (x, y)

def get_x_y(row):
    row['x'], row['y'] = merc(row['lat'], row['lon']) 
    return row

df = pd.read_csv('/home/dataset/Gowalla/loc-gowalla_totalCheckins.txt', 
                 delimiter='\t', 
                 names=['id', 'time', 'lat', 'lon', 'loc_id'], parse_dates=['time'],
                 index_col=['time'])

df['hour'] = df.index.hour
df['day'] = df.index.day
df['new_time'] =df.index.floor('60min')
df['weekday'] = df.index.weekday
#df['week'] = df.index.
df.head(10)
```
